```python
# Filtrando as linhas onde o Municipio seja SALVADOR e Produto seja GASOLINA
ca_df.loc[(ca_df['Municipio'] == 'SALVADOR') & (ca_df['Produto'] == 'GASOLINA')]  
```

```python
""" Criar um dataframe gas_df contendo 
apenas as 4 colunas (Revenda, Municipio, Produto, Valor de Venda)
somente com combustível sendo GASOLINA e exibir na tela """

gas_df = ca_df.loc[ca_df['Produto'] == 'GASOLINA']
display(gas_df)
```

```python
display(gas_df['Valor de Venda'].max())  # Exibindo o maior valor de revenda do DataFrame gas_df
```

```python
#DataFrame.loc[] com múltiplas condições para filtragem
#Quais são os preços, postos que vendem ETANOL na minha cidade (INDAIATUBA) 
#ordenado do menor valor de venda para o maior
etanol_indaiatuba_df = ca_df.loc[(ca_df['Produto'] == 'ETANOL') & (ca_df['Municipio'] == 'INDAIATUBA')]
etanol_indaiatuba_df
```

```python
display(etanol_indaiatuba_df.sort_values(by='Valor de Venda'))  # Ordenando do menor para o maior o valor de vendas 
```

```python
display(etanol_indaiatuba_df.sort_values(by='Valor de Venda', ascending=False))  # Ordenando do maior para o menor o valor de vendas 
```

```python
# Filtrando as linhas onde o Municipio seja SALVADOR e Bairro BARRA e Produto seja GASOLINA OU Municipio seja MANAUS com produto ETANOL 
combustiveis_df.loc[((combustiveis_df['Municipio'] == 'SALVADOR') & (combustiveis_df['Bairro'] == 'BARRA') & (combustiveis_df['Produto'] == 'GASOLINA')) | ((combustiveis_df['Municipio'] == 'MANAUS') & (combustiveis_df['Produto'] == 'ETANOL'))]
```

```python
# Filtrando as linhas onde o Municipio seja SALVADOR e Bairro BARRA e Produto seja GASOLINA OU Municipio seja MANAUS com produto ETANOL, somente as duas primeiras linhas
combustiveis_df.loc[((combustiveis_df['Municipio'] == 'SALVADOR') & (combustiveis_df['Bairro'] == 'BARRA') & (combustiveis_df['Produto'] == 'GASOLINA')) | ((combustiveis_df['Municipio'] == 'MANAUS') & (combustiveis_df['Produto'] == 'ETANOL'))].head(2)
```

```python
# Filtrando as linhas onde o Municipio seja SALVADOR e Bairro BARRA e Produto seja GASOLINA OU Municipio seja MANAUS com produto ETANOL, somente a coluna valor de venda das duas primeiras linhas
combustiveis_df.loc[((combustiveis_df['Municipio'] == 'SALVADOR') & (combustiveis_df['Bairro'] == 'BARRA') & (combustiveis_df['Produto'] == 'GASOLINA')) | ((combustiveis_df['Municipio'] == 'MANAUS') & (combustiveis_df['Produto'] == 'ETANOL'))].head(2)['Valor de Venda']
```

```python
# Filtrando as linhas onde o Municipio seja SALVADOR e Bairro BARRA e Produto seja GASOLINA OU Municipio seja MANAUS com produto ETANOL, o maior valor de venda dentre as duas primeiras linhas
combustiveis_df.loc[((combustiveis_df['Municipio'] == 'SALVADOR') & (combustiveis_df['Bairro'] == 'BARRA') & (combustiveis_df['Produto'] == 'GASOLINA')) | ((combustiveis_df['Municipio'] == 'MANAUS') & (combustiveis_df['Produto'] == 'ETANOL'))].head(2)['Valor de Venda'].max()
```

```python
display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & (combustiveis_df['Municipio'] == 'SAO PAULO') & ((combustiveis_df['Produto'] == 'GASOLINA') | (combustiveis_df['Produto'] == 'GASOLINA ADITIVADA')),['Valor de Venda']].mean())
```

```python
ca_df[['Produto', 'Valor de Venda']]
```

```python
# Menor valor de venda por Produto 
ca_df[['Produto', 'Valor de Venda']].groupby(by='Produto').min()    # groupby sempre é acompanhado de uma função de agregação (max, min, count, sum, mean..)
```

```python
# Maior valor de venda por Produto 
ca_df[['Produto', 'Valor de Venda']].groupby(by='Produto').max().round(1)   # groupby sempre é acompanhado de uma função de agregação (max, min, count, sum, mean..)
```

```python
# Como mostrar média de valor de venda POR COMBUSTÍVEL? 
media_por_combustivel = ca_df[['Produto', 'Valor de Venda']].groupby(by='Produto').mean().round(2)
display(media_por_combustivel)
```
