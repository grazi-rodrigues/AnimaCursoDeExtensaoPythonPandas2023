```python
# Buscando apenas uma coluna
display(leito_ocupacao['municipio'])
```

```python
# Aqui criamos outro DataFrame com apenas algumas colunas 
notificacaoLeito = leito_ocupacao[['dataNotificacao', 'ocupacaoSuspeitoCli', 'ocupacaoSuspeitoUti', 'ocupacaoConfirmadoCli', 'ocupacaoConfirmadoUti', 'ocupacaoCovidUti', 'ocupacaoCovidCli', 'ocupacaoHospitalarUti', 'ocupacaoHospitalarCli', 'saidaSuspeitaObitos', 'saidaSuspeitaAltas', 'saidaConfirmadaObitos', 'saidaConfirmadaAltas', 'estadoNotificacao', 'municipioNotificacao', 'estado', 'municipio']] 
display(notificacaoLeito) 
```

```python
notificacaoLeito.info()
```

```python
notificacaoLeito['estado'].nunique()
```

```python
notificacaoLeito['estado'].unique()     # Houveram notificações nos 26 estados mais o Distrito Federal do Brasil 
```

```python
# Quantidade de municipios que notificaram por estado
notificacaoLeito['municipio'].groupby(by=notificacaoLeito['estado']).nunique()
```

```python
# Validando a contagem
notificacaoLeito[notificacaoLeito['estado'] == 'Minas Gerais']['municipio'].nunique()
```

```python
# Nesta base de dados existem 12 municipios com o mesmo nome em dois estados 
notificacaoLeito['estado'].groupby(by=notificacaoLeito['municipio']).nunique().sort_values(ascending=False).head(20)
```

```python
# Criando a coluna apenas com data de notificação (sem a hora)
notificacaoLeito['data'] = pd.to_datetime(notificacaoLeito['dataNotificacao']).dt.date  
```

```python
# Status de ocupação de leitos por dia 
notificacaoLeito[['ocupacaoSuspeitoCli', 'ocupacaoSuspeitoUti', 'ocupacaoConfirmadoCli', 'ocupacaoConfirmadoUti', 'ocupacaoCovidUti', 'ocupacaoCovidCli', 'ocupacaoHospitalarUti', 'ocupacaoHospitalarCli']].groupby(by=notificacaoLeito['data']).sum().astype(int)
```

```python
# Quantidade de obitos confirmados por dia
notificacaoLeito['saidaConfirmadaObitos'].groupby(by=notificacaoLeito['data']).sum().astype(int)
```
